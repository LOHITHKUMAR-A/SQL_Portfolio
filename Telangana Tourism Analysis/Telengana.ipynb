```python
import pandas as pd
import os
```

```python
# Directory containing the CSV files
csv_dir1 = r"C:\Users\LOHITH KUMAR A\Downloads\Projects\C5 Input for participants\domestic_visitors"
csv_dir2 = r"C:\Users\LOHITH KUMAR A\Downloads\Projects\C5 Input for participants\foreign_visitors"
```

```python
# List all CSV files in the directory
csv_files1 = [file for file in os.listdir(csv_dir1) if file.endswith('.csv')]
csv_files1
```

```python
csv_files2 = [file for file in os.listdir(csv_dir2) if file.endswith('.csv')]
csv_files2
```

```python
# Initialize an empty list to hold DataFrames
df_list1 = []

# Loop through each file and read it into a DataFrame
for file in csv_files1:
    file_path = os.path.join(csv_dir1, file)
    df = pd.read_csv(file_path)
    df_list1.append(df)

# Concatenate all DataFrames
domestic_visitors = pd.concat(df_list1, ignore_index=True)
domestic_visitors
```

```python
df_list2 = []

# Loop through each file and read it into a DataFrame
for file in csv_files2:
    file_path = os.path.join(csv_dir2, file)
    df = pd.read_csv(file_path)
    df_list2.append(df)

# Concatenate all DataFrames
foreign_visitors = pd.concat(df_list2, ignore_index=True)
foreign_visitors
```
